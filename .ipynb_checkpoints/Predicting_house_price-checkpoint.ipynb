#### Binning
Binning can be used to reduce the effects of minor observation errors and make data more readable.
We will use binning for year_built and year_renovated

```python
df_en.columns
```

```python
# Subtract sales year from column 'date':
df_en['Sales_yr'] = df_en['date'].astype(str).str[:4].astype(int)
# Add columns 'age' and 'age_renov':
df_en['age'] = df_en['Sales_yr'] - df_en['yr_built']
df_en['age_renov'] = 0
df_en['age_renov'] = df_en['Sales_yr'][df_en['yr_renovated']!=0] - df_en['yr_renovated'][df_en['yr_renovated']!=0]
# Fill rows without renovation with 0
df_en['age_renov'].fillna(0, inplace=True)
```

```python
# Group age into bins
bins=[-2,5,25,50,75,100,100000]
labels=['<1','1-5','6-25','26-50','51-75','>75']
df_en['age_bin'] = pd.cut(df_en['age'], bins=bins, labels = labels)

# Group age_renov into bins
df_en['age_renov_bin'] = pd.cut(df_en['age_renov'], bins=bins, labels = labels)
```

```python
# Transform the factor values to be able to use in the model
# Drop_first = True to avoid colinearity among dummy variables
df_en = pd.get_dummies(df_en, columns=['age_bin','age_renov_bin'],drop_first = True)
```

```python
# Drop the base columns
df_en = df_en.drop(columns = ['age','age_renov','yr_built','yr_renovated','date'])
```

#### Processing Zipcode Data

More detais on Page 167 of Practical Statistics for Data Scientists by Andrew Bruce, Peter C. Bruce, and Peter Gedeck
- Zipcode is important because it determines location of a house, a deciding factor on sales price.
However, if we use zipcode as raw categorical data, we will increase the number of variables significantly.
In this example, there are 70 distintive values for zipcode, so we will add 70 variables to the model. That would make the model more difficcult to read and fall into the curse of dimensionality.

- The book suggested to group the zip codes according to another variable, such as sale price. Even better is to form zip code groups using the residuals from an initial model.

- I need to do more readings on this.

### Regularization

- Regularization is used to address overfitting and underfitting. 
- Lasso produces simpler and more interpretable models with fewer features because it sinks coefficients to 0. But doesnâ€™t mean more accurate. Lasso works better where a relatively small number of predictions have substantial coefficients, and the rest have small or insignificant coefficients.
- Ridge performs better when all variables have equal coefficients.

- Cross-validation can be used to see which method is better for a particular dataset.

### Polynominal Regression

```python
features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view',
             'grade','yr_built','zipcode']
```

```python
df.columns
```

```python
y=df[['price']]
X=df[features]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7,test_size = 0.3, random_state = 100)
```

```python
polyfeat = PolynomialFeatures(degree=2)
X_allpoly = polyfeat.fit_transform(df[features])
X_trainpoly = polyfeat.fit_transform(X_train)
X_testpoly = polyfeat.fit_transform(X_test)
poly = LinearRegression().fit(X_trainpoly, y_train)
```

```python
pred1 = poly.predict(X_testpoly)
rmsepoly1 = float(format(np.sqrt(metrics.mean_squared_error(y_test,pred1)),'.3f'))
rtrpoly1 = float(format(poly.score(X_trainpoly,y_train),'.3f'))
rtepoly1 = float(format(poly.score(X_testpoly,y_test),'.3f'))
cv1 = float(format(cross_val_score(LinearRegression(),X_allpoly,df['price'],cv=5).mean(),'.3f'))
```

```python
r = evaluation.shape[0]
evaluation.loc[r] = ['Polynomial','All Engineered Features - Degree = 2',rmsepoly1,rtrpoly1,'-',rtepoly1,'-',cv1]
evaluation
```
